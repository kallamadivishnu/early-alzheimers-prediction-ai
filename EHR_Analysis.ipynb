```python
import pandas as pd

ehr_data = {
    "Age": [65, 72, 80],
    "Diabetes": [0, 1, 1],
    "Hypertension": [1, 1, 1],
    "Family_History": [0, 1, 1]
}

df = pd.DataFrame(ehr_data)

print(df)
```

```python
age = 80
diabetes = 1
hypertension = 1
family_history = 1

risk_score = (
    (age / 100) * 0.4 +
    diabetes * 0.2 +
    hypertension * 0.2 +
    family_history * 0.2
)

print("Medical Risk Score:", round(risk_score, 2))

if risk_score > 0.6:
    print("EHR Analysis Result: High Risk")
else:
    print("EHR Analysis Result: Low Risk")
```
