```python
import pandas as pd

cognitive_data = {
    "MMSE_Score": [29, 27, 24, 20, 18],
    "Age": [65, 68, 70, 72, 75]
}

df = pd.DataFrame(cognitive_data)

print(df)
```

```python
def cognitive_risk(mmse):

    if mmse >= 27:
        return "Low Risk"

    elif mmse >= 21:
        return "Moderate Risk"

    else:
        return "High Risk"

df["Risk"] = df["MMSE_Score"].apply(cognitive_risk)

print(df)
```

```python
import matplotlib.pyplot as plt

risk_counts = df["Risk"].value_counts()

plt.figure(figsize=(6,4))
plt.bar(risk_counts.index, risk_counts.values)

plt.title("Cognitive Risk Distribution")
plt.xlabel("Risk Level")
plt.ylabel("Number of Patients")

plt.show()
```

```python
def calculate_risk_score(mmse):

    if mmse >= 27:
        return 0.2

    elif mmse >= 21:
        return 0.6

    else:
        return 0.9

df["Risk_Score"] = df["MMSE_Score"].apply(calculate_risk_score)

print(df)
```
