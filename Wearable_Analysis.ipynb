```python
import pandas as pd

wearable_data = {
    "Heart_Rate": [72, 85, 95, 110],
    "Sleep_Hours": [8, 7, 6, 4],
    "Activity_Level": [9000, 7000, 5000, 2000]
}

df = pd.DataFrame(wearable_data)

print(df)
```

```python
heart_rate = 110
sleep_hours = 4
activity_level = 2000

wearable_risk = (
    (heart_rate / 120) * 0.4 +
    ((8 - sleep_hours) / 8) * 0.3 +
    ((10000 - activity_level) / 10000) * 0.3
)

print("Wearable Risk Score:", round(wearable_risk, 2))

if wearable_risk > 0.6:
    print("Wearable Analysis Result: High Risk")
else:
    print("Wearable Analysis Result: Low Risk")
```

```python
import matplotlib.pyplot as plt

metrics = [
    "Heart Rate",
    "Sleep",
    "Activity"
]

values = [
    heart_rate / 120,
    (8 - sleep_hours) / 8,
    (10000 - activity_level) / 10000
]

plt.figure(figsize=(6,4))
plt.bar(metrics, values)

plt.title("Wearable Risk Factors")
plt.ylabel("Risk Contribution")

plt.show()
```
