```python
print("Multimodal AI Framework for Early Alzheimer's Disease Prediction")
```

```python
mri_risk = 0.72
speech_risk = 0.61
emotion_risk = 0.69
cognitive_risk = 0.90
ehr_risk = 0.92
wearable_risk = 0.76

print("MRI Risk Score       :", mri_risk)
print("Speech Risk Score    :", speech_risk)
print("Emotion Risk Score   :", emotion_risk)
print("Cognitive Risk Score :", cognitive_risk)
print("EHR Risk Score       :", ehr_risk)
print("Wearable Risk Score  :", wearable_risk)
```

```python
final_risk = (
    mri_risk +
    speech_risk +
    emotion_risk +
    cognitive_risk +
    ehr_risk +
    wearable_risk
) / 6

print("Final Alzheimer Risk Score:", round(final_risk, 2))
```

```python
if final_risk < 0.4:
    prediction = "LOW RISK"
elif final_risk < 0.7:
    prediction = "MODERATE RISK"
else:
    prediction = "HIGH RISK"

print("Prediction:", prediction)
```

```python
import matplotlib.pyplot as plt

modules = [
    "MRI",
    "Speech",
    "Emotion",
    "Cognitive",
    "EHR",
    "Wearable"
]

scores = [
    mri_risk,
    speech_risk,
    emotion_risk,
    cognitive_risk,
    ehr_risk,
    wearable_risk
]

plt.figure(figsize=(8,5))

plt.bar(modules, scores)

plt.title("Multimodal Alzheimer Risk Analysis")
plt.ylabel("Risk Score")

plt.show()
```

```python
print("="*60)
print("ALZHEIMER EARLY PREDICTION REPORT")
print("="*60)

print("MRI Risk Score       :", mri_risk)
print("Speech Risk Score    :", speech_risk)
print("Emotion Risk Score   :", emotion_risk)
print("Cognitive Risk Score :", cognitive_risk)
print("EHR Risk Score       :", ehr_risk)
print("Wearable Risk Score  :", wearable_risk)

print("\nFinal Risk Score:", round(final_risk, 2))

print("Prediction:", prediction)

if prediction == "LOW RISK":
    print("\nRecommendation: Maintain healthy lifestyle and regular monitoring.")
elif prediction == "MODERATE RISK":
    print("\nRecommendation: Periodic cognitive assessment is advised.")
else:
    print("\nRecommendation: Immediate neurological evaluation recommended.")

print("="*60)
```
